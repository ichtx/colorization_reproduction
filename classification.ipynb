```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.applications.vgg16 import VGG16
from keras.utils import load_img
from keras.utils import img_to_array
from keras.utils import image_dataset_from_directory
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from sklearn.metrics import accuracy_score
import json
```

```python
file = open("Imagenet100/Labels.json")
classname = json.load(file)
symbol = list(classname.keys())
symbol = sorted(symbol)
print(symbol)
```

```python
model = VGG16(weights="imagenet")

folder = "ECCV/validation"

img_val = image_dataset_from_directory(folder, batch_size=None, image_size=(224,224), shuffle=False)
data = np.array(list(map(lambda x: x[0], img_val)))
label = np.array(list(map(lambda x: x[1], img_val)))
```

```python
predicted = model.predict(data)
predlabel = decode_predictions(predicted)
```

```python
pred = list(map(lambda x: x[0][0], predlabel))
real = list(map(lambda x: symbol[x], label))
acc = accuracy_score(real, pred)
print(acc)
```

```python
originalfolder = "Imagenet100/validation"

orginalimg = image_dataset_from_directory(originalfolder, batch_size=None, image_size=(224,224), shuffle=False)
originaldata = np.array(list(map(lambda x: x[0], orginalimg)))
originallabel = np.array(list(map(lambda x: x[1], orginalimg)))
```

```python
imagenetpred = model.predict(originaldata)
imagenetlabel = decode_predictions(imagenetpred)
```

```python
originalpred = list(map(lambda x: x[0][0], imagenetlabel))
# real = list(map(lambda x: symbol[x], label))
acc = accuracy_score(real, originalpred)
print(acc)
```

```python
with open ("imagenetpred.txt", "w") as file:
    for item in imagenetlabel:
        file.write("%s" % item)
        file.write("\n")
file.close()
```
